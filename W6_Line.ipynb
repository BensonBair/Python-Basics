# Matplotlib (http://matplotlib.org/):  Python的2D繪圖程式庫

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
plt.plot([1, 2, 3, 4, 5]) # x-axis ranges from 0-4 and the y-axis from 1-5
# 只給定一個 single list 或 array，就是y軸，x軸從0開始。
plt.show() # 將圖呈現出來
```

```python
plt.plot([1, 2, 3, 4, 5], [1, 4, 9, 16, 25]) # 預設是 b-：藍色實線
plt.show()
```

```python
plt.plot([1, 2, 3, 4, 5], [1, 4, 9, 16, 25], "ro") # 第三個引數可以指定 顏色 樣式 。"ro": 紅色圓點
plt.show()
```

```python
plt.plot([1, 2, 3, 4, 5], [1, 4, 9, 16, 25], "ro") 
plt.axis([0, 10, 0, 30]) # 指定軸的範圍：a list of [xmin, xmax, ymin, ymax]
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0., 10., 0.5)
# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(1,11)
y = np.arange(0, 50, 5)
plt.plot(x, y, linewidth=5.0) #設定線的寬度
plt.show()
```

```python
x1 = np.arange(1,11)
y1 = x1**2
x2 = x1
y2 = x2**3
lines = plt.plot(x1, y1, x2, y2)
plt.setp(lines, color='r', linewidth=3.0) # 用關鍵字引數設定屬性
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1) # figure()：第幾張圖
plt.subplot(211) # 將圖切割成2列，每列1個子圖，在第1個子圖位置
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212) # 將圖切割成2列，每列1個子圖，在第2個子圖位置
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(1)                # 第一張圖 
plt.subplot(211)             # 第一張圖的第1張子圖
plt.plot([1, 2, 3])
plt.subplot(212)             # 第一張圖得第2張子圖
plt.plot([4, 5, 6])


plt.figure(2)                # 第二張圖
plt.plot([4, 5, 6])          # 預設就是 subplot(111) 

plt.figure(1)                # 第一張圖。current 是 子圖subplot(212) 
plt.subplot(211)             # 將 子圖subplot(211) 設為 第一張圖 的 current
plt.title('Easy as 1, 2, 3') # 子圖subplot(211)的標題
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

x = np.arange(0.0, 5.0, 0.1)
plt.plot(x, f(x))

plt.xlabel("X label", fontsize=14, color='red')   # x軸標籤 
plt.ylabel("Y label", fontsize=14, color='green') # y軸標籤
plt.title("Plot of x vs. y") # 圖標題
plt.text(1.5, -0.4, "put text here", fontsize=14, color="red") #文字說明
plt.annotate('local max', xy=(1, 0.4), xytext=(2.5, 0.8), fontsize=16,
             arrowprops=dict(facecolor='blue')) # 加上註解
plt.grid(True) #是否加上格線
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 1, 500) # 從0到1中間取500個sample 
y = np.sin(4 * np.pi * x) * np.exp(-5 * x) 

fig, ax = plt.subplots() # fig : matplotlib.figure.Figure object, ax : Axes object 

ax.fill(x, y, "g", zorder=10) # fill填滿範圍
ax.grid(True, zorder=5) # zorder 數字小的先畫(在下層)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 500) # Return evenly spaced numbers over a specified interval.
y1 = np.sin(x)
y2 = np.sin(3 * x)

fig, ax = plt.subplots()
ax.fill(x, y1, 'b', x, y2, 'r', alpha=0.3) # alpha是透明程度 (從0到1)
plt.show()
```
