# Scatter Plot (散佈圖)

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

N = 50
x = np.random.rand(N) #rand()隨機產生 N個0.0到1.0之間的值
y = np.random.rand(N) 
plt.scatter(x, y) # 散佈圖 (x, y) Input data  : array_like, shape (n, ) 
#plt.show() # 將圖呈現
plt.savefig(filename = "my_scatter.png", format = "png")
```

```python
area = np.pi * (15 * np.random.rand(N))**2  # 半徑r從0到15，圓的面積 pi*pow(r,2) 
plt.scatter(x, y, s=area)
plt.show()
```

```python
colors = np.random.rand(N) # an array of float
plt.scatter(x, y, s=area, c=colors)
plt.show()
```

```python
colors = np.random.rand(N)
plt.scatter(x, y, s=area, c=colors, alpha=0.5) #alpha 是透明程度 從 0(transparent) 到 1(opaque)
plt.show()
```

```python
colors = np.random.rand(N)
plt.scatter(x, y, s=area, c=colors, alpha=0.7, marker="*")
# marker-->description： "."-->point、","-->pixel、"o"-->circle
#"v"-->triangle_down、"^"-->triangle_up、"<"-->triangle_left、">"-->triangle_right
#"1"-->tri_down、"2"-->tri_up、"3"-->tri_left、"4"-->tri_right、"8"-->octagon
#"s"-->square、"p"-->pentagon、"P"-->plus (filled)、"*"-->star
#"h"-->hexagon1、"H"-->hexagon2、"+"-->plus、"x"-->x、"X"-->x (filled)
#"D"-->diamond、"d"-->thin_diamond、"|"-->vline、"_"-->hline
plt.show()
```

```python
colors = np.random.rand(N)
plt.xlabel("x values") # x軸說明
plt.ylabel("y values") # y軸說明
plt.scatter(x, y, s=area, c=colors, alpha=0.9, marker="3")
plt.show()
```

```python
colors = np.random.rand(N)
plt.rcParams["font.sans-serif"]=["SimHei"] #用来正常顯示中文
plt.title("圖標題", fontsize=18) 
plt.scatter(x, y, s=area, c=colors, alpha=0.4, marker="D")
plt.show()
```

```python
import matplotlib.pyplot as plt
from numpy.random import rand

plt.figure(figsize=(15,5)) # 改變圖的大小，單位是寸
for color in ["red", "green", "blue"]:
    n = 250
    x, y = rand(2, n)
    scale = 200.0 * rand(n)
    plt.scatter(x, y, c=color, s=scale, label=color,
               alpha=0.5, edgecolors="none") # 點沒有邊的顏色
plt.legend(fontsize="xx-large") # 圖例說明
plt.grid(True) #  加上格線
plt.show()
```
