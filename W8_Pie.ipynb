顏色命名範例

```python
"""
========================
Visualizing named colors
========================

Simple plot example with the named colors and its visual representation.
"""
from __future__ import division

import matplotlib.pyplot as plt
from matplotlib import colors as mcolors


colors = dict(mcolors.BASE_COLORS, **mcolors.CSS4_COLORS)

# Sort colors by hue, saturation, value and name.
by_hsv = sorted((tuple(mcolors.rgb_to_hsv(mcolors.to_rgba(color)[:3])), name)
                for name, color in colors.items())
sorted_names = [name for hsv, name in by_hsv]

n = len(sorted_names)
ncols = 4
nrows = n // ncols + 1

fig, ax = plt.subplots(figsize=(16, 10))

# Get height and width
X, Y = fig.get_dpi() * fig.get_size_inches()
h = Y / (nrows + 1)
w = X / ncols

for i, name in enumerate(sorted_names):
    col = i % ncols
    row = i // ncols
    y = Y - (row * h) - h

    xi_line = w * (col + 0.05)
    xf_line = w * (col + 0.25)
    xi_text = w * (col + 0.3)

    ax.text(xi_text, y, name, fontsize=(h * 0.8),
            horizontalalignment='left',
            verticalalignment='center')

    ax.hlines(y + h * 0.1, xi_line, xf_line,
              color=colors[name], linewidth=(h * 0.6))

ax.set_xlim(0, X)
ax.set_ylim(0, Y)
ax.set_axis_off()

fig.subplots_adjust(left=0, right=1,
                    top=1, bottom=0,
                    hspace=0, wspace=0)
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.figure(figsize=(5,5)) #設定圖的大小，figsize(寬,高)，單位是寸
sizes = [60, 30, 10] 
c = ['lightcoral','yellowgreen','darkblue'] 
e = (0.05,0,0) # 將每個部分給分隔出來，數值部分是與其他非分隔部分的間隙
plt.pie(sizes, colors = c, explode=e)
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.figure(figsize=(6,5))
sizes = [60, 30, 10] 
c = ['red','yellowgreen','lightskyblue'] 
e = (0.05,0,0)
tag = ["第一部分","第二部分","第三部分"] #圓餅圖中，每個部分對應的標籤
plt.rcParams["font.sans-serif"]=["SimHei"] #用来正常顯示中文
# Windows的字體對應名稱
# 黑體：SimHei, 微軟雅黑：Microsoft YaHei, 微軟正黑體：Microsoft JhengHei, 新宋體：NSimSun
# 新細明體：PMingLiU，細明體：MingLiU, 標楷體：DFKai-SB, 仿宋：FangSong, 楷體：KaiTi
plt.pie(sizes, colors = c, explode=e, labels = tag, labeldistance=1.2)
# labeldistance 是指標籤離圓點的距離，1.2是指1.2倍半徑的位置
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.figure(figsize=(6,5))
sizes = [60, 30, 10] 
c = ['red','yellowgreen','lightskyblue']
e = (0.05,0,0)
tag = ["第一部分","第二部分","第三部分"] 
plt.rcParams["font.sans-serif"]=["SimHei"] 
plt.pie (sizes, colors = c, explode=e, labels=tag , labeldistance = 1.2,
        autopct = '%3.1f%%', pctdistance = 0.6, shadow = False, startangle = 90)
# autopct：圓裡面的百分比格式，%3.1f%%表示小數有3位，整數有1位的浮點數
# pctdistance：百分比離圓心的距離
# shadow：是否有陰影
# startangle：起始角度，預設是0，表示從0開始逆時針轉是第一部分。一般從90度開始會比較好看 
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.figure(figsize=(9,6))
sizes = [60, 30, 10] 
c = ['red','yellowgreen','lightskyblue']
e = (0.05,0,0)
tag = ["第一部分","第二部分","第三部分"] 
plt.rcParams["font.sans-serif"]=["SimHei"] 
patches,l_text,p_text = plt.pie (sizes, colors = c, explode=e, labels=tag , labeldistance = 1.2,
                       autopct = '%3.1f%%', pctdistance = 0.6, shadow = False, startangle = 90)
#如果 autopct 不是none，回傳 the tuple (patches, texts, autotexts)
# patches is a sequence of matplotlib.patches.Wedge instances
# texts is a list of the label matplotlib.text.Text instances 標籤文字串列
# autotexts is a list of Text instances for the numeric labels 百分比文字串列

#用set_size()函式改變文字的大小
for t in l_text:
    t.set_size=(30)
for t in p_text:
    t.set_size=(20)

plt.axis("equal") #設定 x軸 與 y軸的刻度一致，這樣圓餅圖才能是圓的
plt.legend() # 加上圖例說明 
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # 將第2部份切割出來 

fig1, ax1 = plt.subplots() # Number of rows/columns of the subplot grid. 預設是1
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90) # 加上陰影
ax1.axis('equal') 

plt.show()
```
