# Bar Plot (長條圖)

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.bar([1, 2, 3, 4, 5], [5, 9, 3, 7, 4]) #預設長條寬度是0.8，顏色是藍色
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

#並列的長條圖
x = np.arange(1,6)
plt.bar(x-0.2, [5, 9, 3, 7, 4], width=0.4) #長條寬度設為0.8
plt.bar(x+0.2, [2, 5, 10, 6, 8], width=0.4, color="red") #顏色設為紅色
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1,6)
A = np.array([5, 9, 3, 7, 4])
B = np.array([2, 5, 10, 6, 8])
C = np.array([8, 6, 4, 3, 2])

#將長條圖疊起來
plt.bar(x, A, width=0.5)
plt.bar(x, B, color="red", width=0.5, bottom = A) # 疊在A長條上面
plt.bar(x, C, color="green", width=0.5, bottom = A+B) # 疊在A+B長條的上面
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1,6)
# barh()：水平長條圖
plt.barh(x, [5, 9, 3, 7, 4])
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1,6)
A = np.array([5, 9, 3, 7, 4])
B = np.array([2, 5, 10, 6, 8])

# 比較兩組資料
plt.barh(x, A)
plt.barh(x, -B, color="red") # 另一組加上負號，變成反方向
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcdefaults() # 恢復預設參數值
fig, ax = plt.subplots()
people = ('Tom', 'Dick', 'Harry', 'Slim', 'Jim')
y_pos = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people)) # 誤差值
ax.barh(y_pos, performance, xerr=error, align="edge", color="green", ecolor="black") 
# center 長條置中對齊 / egde 水平長條置於下方、垂直長條至於左側
ax.set_yticks(y_pos)
ax.set_yticklabels(people)
ax.invert_yaxis() # 標籤由上向下讀  
ax.set_xlabel('Performance')
ax.set_title('How fast do you want to go today?')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = 5
men_means = (20, 35, 30, 35, 27)
men_std = (2, 3, 4, 1, 2)

ind = np.arange(N)  # the x locations for the groups
width = 0.35       # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(ind, men_means, width, color='r', yerr=men_std)

women_means = (25, 32, 34, 20, 25)
women_std = (3, 5, 2, 3, 3)
rects2 = ax.bar(ind + width, women_means, width, color='y', yerr=women_std)

# add some text for labels, title and axes ticks
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(('G1', 'G2', 'G3', 'G4', 'G5'))

#ax.legend((rects1[0], rects2[0]), ('Men', 'Women'))


def autolabel(rects):
    """
    Attach a text label above each bar displaying its height
    """
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.05*height,
                '%d' % int(height),
                ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

plt.show()
```
